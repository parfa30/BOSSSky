```python
def opening_angle(ra1, dec1, ra2, dec2):
    # Compute the opening angles between (ra1,dec1) and (ra2,dec2)
    # in degrees using the Haversine formula.
    ra1, ra2 = np.deg2rad(ra1), np.deg2rad(ra2)
    dec1, dec2 = np.deg2rad(dec1), np.deg2rad(dec2)
    havRA12 = 0.5 * (1 - np.cos(ra2 - ra1))
    havDEC12 = 0.5 * (1 - np.cos(dec2 - dec1))
    havPHI = havDEC12 + np.cos(dec1) * np.cos(dec2) * havRA12
    return np.rad2deg(np.arccos(np.clip(1 - 2 * havPHI, -1, +1)))
```

```python
def unitvec(ra, dec):
    # Return a unit vector aligned with (ra,dec) in degrees.
    ra, dec = np.deg2rad(ra), np.deg2rad(dec)
    cosdec = np.cos(dec)
    return np.array([cosdec * np.cos(ra), cosdec * np.sin(ra), np.sin(dec)])
```

```python
def at_horizon(raz, decz, ra, dec):
    # Project (ra,dec) into the horizon plane defined by the zenith (raz,decz).
    # Invalid when (ra,dec) points along +/-zenith.
    z = unitvec(raz, decz)
    v = unitvec(ra, dec)
    dot = (z * v).sum(axis=0)
    assert np.all(dot < 1), '(ra,dec) points along +/-zenith'
    return (v - dot * z) / np.sqrt(1 - dot ** 2)
```

```python
def load_data(name='allsky2.fits'):
    data = astropy.table.Table.read(name)
    # Compute the moon-pointing opening angle in degrees.
    moon_sep = opening_angle(data['RA'], data['DEC'], data['MOON_RA'], data['MOON_DEC'])
    data['MOON_SEP'] = astropy.table.Column(moon_sep.astype(np.float32), unit='deg')
    # Compute the moon zenith angle in degrees.
    moon_zenith = opening_angle(data['ZEN_RA'], data['ZEN_DEC'],
                                data['MOON_RA'], data['MOON_DEC'])
    data['MOON_ZENITH'] = astropy.table.Column(moon_zenith.astype(np.float32), unit='deg')
    # Compute the pointing zenith angle in degrees.
    obs_zenith = opening_angle(data['ZEN_RA'], data['ZEN_DEC'],
                               data['RA'], data['DEC'])
    data['OBS_ZENITH'] = astropy.table.Column(obs_zenith.astype(np.float32), unit='deg')
    # Convert illuminated fraction to the phase convention of the KS model,
    # where 0 = full, 1 = new.
    mphase = np.arccos(2 * data['MOON_FRAC'] - 1) / np.pi
    data['MPHASE'] = astropy.table.Column(mphase.astype(np.float32))
    # Compute the predicted V-band magnitude of scattered moonlight.
    moonV = specsim.atmosphere.krisciunas_schaefer(
        obs_zenith * u.deg, moon_zenith * u.deg, moon_sep * u.deg, mphase,
        vband_extinction=0.15154).value
    moon_up = (moon_zenith < 90) & (data['MOON_PHASE'] > 0)
    moonV[~moon_up] = 0.
    data['MOONV'] = astropy.table.Column(moonV)
    # Compute the pointing-sun relative azimuth angle in the range [0,180] deg.
    obs_horizon = at_horizon(data['ZEN_RA'], data['ZEN_DEC'],
                             data['RA'], data['DEC'])
    sun_horizon = at_horizon(data['ZEN_RA'], data['ZEN_DEC'],
                             data['SUN_RA'], data['SUN_DEC'])
    dot = np.clip((obs_horizon * sun_horizon).sum(axis=0), -1., +1.)
    sun_relaz = np.rad2deg(np.arccos(dot))
    data['SUN_RELAZ'] = astropy.table.Column(sun_relaz.astype(np.float32), unit='deg')

    return data

data = load_data()
```

```python
def plot_corner(data=data):
    good = np.isfinite(data['AIRMASS'])
    print('Keeping {} / {} rows with finite AIRMASS'.format(np.count_nonzero(good), len(data)))
    fields = ['SEEING50', 'AIRMASS', 'GSKYFLUX', 'SUN_ANGLE', 'MOON_FRAC', 'MOON_ANGLE', 'MOON_SEP']
    plot_data = np.array(data[fields][good])
    print(plot_data.dtype.names)
    flat = plot_data.view((np.float32, len(plot_data.dtype.names)))
    corner.corner(flat, labels=plot_data.dtype.names)
    
plot_corner()
```

```python
def plot_dark(data=data, xname='AIRMASS', yname='GSKYFLUX', vname='SEEING50'):
    x = data[xname]
    y = data[yname]
    v = data[vname]
    VALID = np.isfinite(x) & np.isfinite(y)
    SEL = (data['SUN_ANGLE'] < -20) & (data['MOON_ANGLE'] < -5) & VALID
    xmin, xmax = np.percentile(x[SEL], (1, 99))
    ymin, ymax = np.percentile(y[SEL], (1, 99))
    vmin, vmax = np.percentile(v[SEL], (1, 99))
    
    fig, ax = plt.subplots(2, 2, figsize=(8, 8))
    ax = ax.flatten()
    
    s = ax[0].scatter(x[SEL], y[SEL], s=2, lw=0, c=v[SEL], vmin=vmin, vmax=vmax)
    plt.colorbar(s, ax=ax[0]).set_label(vname)
    
    x_model = np.linspace(xmin, xmax, 50)
    model = sklearn.linear_model.LinearRegression(fit_intercept=False)
    getX = lambda x: np.vstack((1 + 0*x, x)).transpose()
    model.fit(getX(x[SEL]), y[SEL])
    print(list(model.coef_))
    y_model = model.predict(getX(x_model))
    ax[0].plot(x_model, y_model, 'r--')

    ax[0].set_xlim(xmin, xmax)
    ax[0].set_ylim(ymin, ymax)
    ax[0].set_xlabel(xname)
    ax[0].set_ylabel(yname)
    
    ax[1].hist(y, range=(ymin, ymax), bins=50, histtype='step', color='b')
    
    ypred = model.predict(getX(x[SEL]))
    yres = y[SEL] - ypred + np.median(y) #model.predict(getX(1. + 0*x[SEL]))
    ax[1].hist(yres, range=(ymin, ymax), bins=50, histtype='step', color='r')
    
    s = ax[2].scatter(x[SEL], yres, s=2, lw=0, c=v[SEL], vmin=vmin, vmax=vmax)
    plt.colorbar(s, ax=ax[2]).set_label(vname)
    ax[2].set_xlim(xmin, xmax)
    ax[2].set_ylim(ymin, ymax)
    ax[2].set_xlabel(xname)
    ax[2].set_ylabel(yname)
    
    ax[3].axis('off')

plot_dark()
```

```python
def plot_twilight(data=data, xname='SUN_ANGLE', yname='GSKYFLUX', vname='AIRMASS'):
    x = data[xname]
    y = data[yname]
    v = data[vname]
    VALID = np.isfinite(x) & np.isfinite(y)
    SEL = (data['SUN_ANGLE'] > -22) & (data['MOON_ANGLE'] < -5) & VALID
    xmin, xmax = np.percentile(x[SEL], (1, 99))
    ymin, ymax = np.percentile(y[SEL], (1, 99))
    vmin, vmax = np.percentile(v[SEL], (1, 99))
    
    fig, ax = plt.subplots(2, 2, figsize=(8, 8))
    ax = ax.flatten()

    s = ax[0].scatter(x[SEL], y[SEL], s=5, lw=0, c=v[SEL], vmin=vmin, vmax=vmax)
    plt.colorbar(s, ax=ax[0]).set_label(vname)
    
    x_model = np.linspace(xmin, xmax, 50)
    model = sklearn.linear_model.LinearRegression(fit_intercept=False)
    getX = lambda x: np.vstack((1 + 0*x, 2. ** (18.5 + x), 3. ** (18.5 + x),
                                4. ** (18.5 + x))).transpose()
    model.fit(getX(x[SEL]), y[SEL])
    print(list(model.coef_))
    y_model = model.predict(getX(x_model))
    ax[0].plot(x_model, y_model, 'r--')
    
    y_djs = 1 + 3 ** (18.5 + x_model) / 30.
    ax[0].plot(x_model, y_djs, 'b-')

    ax[0].set_xlim(xmin, xmax)
    ax[0].set_ylim(ymin, ymax)
    ax[0].set_xlabel(xname)
    ax[0].set_ylabel(yname)

plot_twilight()
```

```python
def plot_moon(data=data, xname='MOONV', yname='GSKYFLUX', vname='MOON_ANGLE'):
    x = data[xname]
    y = data[yname]
    v = data[vname]
    airmass = data['AIRMASS']
    VALID = np.isfinite(x) & np.isfinite(y) & np.isfinite(airmass)
    SEL = (data['SUN_ANGLE'] < -18) & (data['MOONV'] > 0) & (data['MOON_ANGLE'] > 5) & VALID

    # Subtract the predicted dark sky component.
    dark_coefs = np.array([0.29583195, 0.62750822])
    X0 = 1.
    y0 = dark_coefs[0] + dark_coefs[1] * X0
    dark_pred = dark_coefs[0] + dark_coefs[1] * airmass
    y = (y - dark_pred + y0) / y0
    
    y = -2.5 * np.log10(y)
        
    # Calculate percentile limits.
    xmin, xmax = np.percentile(x[SEL], (1, 99))
    ymin, ymax = np.percentile(y[SEL], (1, 99))
    vmin, vmax = np.percentile(v[SEL], (1, 99))

    fig, ax = plt.subplots(2, 2, figsize=(8, 8))
    ax = ax.flatten()

    s = ax[0].scatter(x[SEL], y[SEL], s=5, lw=0, c=v[SEL], vmin=vmin, vmax=vmax)
    plt.colorbar(s, ax=ax[0]).set_label(vname)
    ax[0].axhline(1, c='r', ls='-')
    
    '''
    x_model = np.linspace(xmin, xmax, 50)
    model = sklearn.linear_model.LinearRegression(fit_intercept=False)
    getX = lambda x: np.vstack((1 + 0*x, x * np.exp(x))).transpose()
    model.fit(getX(x[SEL]), y[SEL])
    print(list(model.coef_))
    y_model = model.predict(getX(x_model))
    plt.plot(x_model, y_model, 'r--')
    
    y_djs = 1 + 4000 ** x_model / 30.
    plt.plot(x_model, y_djs, 'b-')
    '''

    ax[0].set_xlim(xmin, xmax)
    ax[0].set_ylim(ymin, ymax)
    ax[0].set_xlabel(xname)
    ax[0].set_ylabel('(SKY - DARK(X) + DARK(X=1)) / DARK(X=1)')
    
    mp = data['MOON_PHASE'][SEL]
    gsky = data['GSKYFLUX'][SEL]
    s = ax[1].scatter(mp, gsky,
                      s=5, lw=0, c=v[SEL], vmin=vmin, vmax=vmax)
    ax[1].set_xlim(np.percentile(mp, (1, 99)))
    ax[1].set_ylim(np.percentile(gsky, (1, 99)))
```
